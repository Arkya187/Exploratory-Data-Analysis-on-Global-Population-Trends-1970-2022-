```python
# Selecting only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Computing correlation matrix
corr_matrix = df_numeric.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Concatenating 'Continent' back to allow groupby
df_numeric_with_continent = pd.concat([df['Continent'], df_numeric], axis=1)

# Group by Continent and computing mean
continent_grouped = df_numeric_with_continent.groupby('Continent').mean()

# Sort by '2022 Population'
result = continent_grouped.sort_values(by="2022 Population", ascending=False)

print(result)
```
